Matplotlib es una biblioteca completa para crear visualizaciones estáticas, animadas e interactivas en Python. Matplotlib hace que las cosas fáciles sean fáciles y las difíciles sean posibles.

# Pyplot básico

Pyplot es una herramienta que tiene Matplotlib para ejecutar gráficas de manera sencilla

Para usarla usualmente la importamos de la siguiente manera:

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Creando un gráfico de muestra

```python
# Creamos las variables
x = np.linspace(0, 5, 11)
y = x**2
# Creamos la gráfica de y respecto a x
plt.plot(x, y)
# Mostramos la gráfica
plt.show()
```

## Modificando el gráfico

### Cambiando el color

Podemos cambiar el color de nuestro gráfico agregando una tercer entrada con un string que contenga la inicial de el color deseado:

| entrada | color        |
|---------|--------------|
| 'b'     | azul(blue)   |
| 'g'     | verde(green) |
| 'm'     | magenta      |

Estos son solo unos ejemplos

```python
plt.plot(x, y, 'r')
plt.show()
```

### Agregando marcadores

Podemos agregar marcadores igualmente agregando el estilo en la tercer entrada, por ejemplo:

| Entrada | Tipo de marcador       |
|---------|------------------------|
| '.'     | punto                  |
| ','     | pixel                  |
| 'o'     | circular               |
| 'v'     | triángulo hacia abajo  |
| '^'     | triángulo hacia arriba |
| 's'     | cuadrado               |
| '*'     | estrella               |

```python
plt.plot(x, y, '*')
plt.show()
```

### Cambiando el estilo de línea

Podemos cambiar también el estilo de la línea, esto de manera similar siguiendo la sintaxis:

| Entrada | Estilo                    |
|---------|---------------------------|
| '-'     | linea solida              |
| '--'    | línea discontinua         |
| '-.'    | línea de puntos y guiones |
| ':'     | línea punteada            |

```python
plt.plot(x, y, '-.')
plt.show()
```

> Podemos combinarlos agregando directamente uno después de otro en la tercer entrada (al estilo de Matlab)

```python
plt.plot(x, y, 'ro-.')
plt.show()
```

## Otros tipos de gráficos

Podemos crear otros tipos de gráficos, para ello bastará con reemplazar `plot` por el tipo de gráfico deseado

### Histograma

```python
plt.hist(y)
plt.show()
```

### Gráfica circular

```python
plt.pie(y)
plt.show()
```

# Subplot

Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a través de índices:

```python
# Creando una matriz de gráficos con una fila y dos columnas

plt.subplot(1,2,1) # Gráfico 1
plt.plot(x,y, 'r--') 

plt.subplot(1,2,2)  # Gráfico 2
plt.pie(y)
plt.show()
```

Podemos seguir trabajando con cada subplot

```python
# Agregando otro plot dentro de un subplot

plt.subplot(1,2,1)
plt.plot(x,y, 'r--') # Plot 1
plt.plot(y,x, 'b:') #Plot 2

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

> Podemos ordenar nuestros gráficos de la forma mas conveniente para nosotros

```python
# Tres gráficos

plt.subplot(2,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.scatter(y,x,)
plt.subplot(2,2,3)
plt.hist(y)
plt.show()
```

# Método orientado a objetos

Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gráficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gráficas las cuales contienen diferentes ejes.

| Pyplot          | Object Oriented                    |
|-----------------|------------------------------------|
| Rápido          | Mayor personalización              |
| Fácil           | Más código                         |
| Una sola figura | Más amigable a múltiples diagramas |

Utilizamos el concepto de objeto para crear una gráfica más personalizable

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'b')
plt.show()
```

Podemos crear varias gráficas dentro del lienzo llamando el método fig.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()
```

> Dado que los axes son objetos independientes, podemos trabajar directamente con ellos

# Subplots

Creamos un gráfico tipo fig.

```python
fig, axes = plt.subplots()
axes.plot(x,y)
plt.show()
```

Cuando sea necesario presentar varios gráficos en una sola fig. Bastara con agregar el numero de filas y columnas, y colocar la gráfica en la posición deseada

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

Otra forma es mediante el nombre almacenado en una tupla

```python
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2)
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
plt.show()
```

Podemos incluso generar una matriz de gráficos que sera rellenada con gráficos vacíos

```python
fig, axes = plt.subplots(2,4)

#genera un trazo accediendo a las gráficas a traves del indice de la matriz
axes[0,0].plot(x,y)
axes[0,1].plot(y,x, 'r')

fig.tight_layout() #mejora la visualización de los ejes de cada gráfico
```

Alternativa para generar una matriz de tipo fig, accediendo a cada gráfica mediante los nombres dados en una tupla

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2,4)

ax1.plot(x, np.sin(x), 'b')
ax2.plot(y, np.cos(x), 'purple')
ax3.plot(x, np.tan(y), 'orange')
ax4.plot(y, np.cos(y)**2, 'black')

#mejorar la visualización de las figuras
fig.tight_layout()
```

```python
# Ejemplo completo

fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)

ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'r')
ax3.plot(x,np.tan(x),'y')
ax4.plot(x,np.cos(x)**2)

ax5.plot(x,np.cos(x)**(-1),color=('#8B008B'))
ax6.plot(x,np.sin(x)**(-1),color=('#F08080'))
ax7.plot(x,np.tan(x)**(-1),color=('chocolate'))
ax8.plot(x,np.cos(x)**3,color=('magenta'))

fig.tight_layout()
```

# Leyendas, etiquetas, títulos, tamaño

Al momento de presentar nuestra información en gráficos es completamente necesario agregar un contexto a las mismas

```python
y = np.sin(x)

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
plt.show()
```

## Agregar títulos

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax1.set_title("Relación X - Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relación Y - X")
```

## Agregando etiquetas de los ejes

```python
fig, (ax1,ax2) = plt.subplots(1,2)

ax1.plot(x,y, 'b', )
ax1.set_title("Relación X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relación Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
```

## Agregando leyendas

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title("Relación X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
```

*podemos cambiar la localización de la leyenda*

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "$sin(x)$") # utilizamos notación de latex $sin()$ para funciones matemáticas 
ax1.set_title("Relación X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend(loc = "center")
```

> Podemos también modificar el tamaño de las gráficas

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (1,7))
```

## Personalizar colores y estilos en Pyplot

Para personalizar con pyplot es un poco más largo cuando programamos, pero se puede personalizar muchísimo más

| Modificación           | Entrada         |
|------------------------|-----------------|
| Transparencia de linea | alpha           |
| Grosor de linea        | linewidth       |
| Estilo de linea        | linestyle       |
| Estilo de marcador     | marker          |
| Tamaño de marcador     | markersize      |
| Color de marcador      | markerfacecolor |

```python
fig, ax = plt.subplots(figsize = (6,6))
plt.style.use('dark_background')

ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4, color = 'white',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
```

# Bar plot

El gráfico de barras nos permite gráficar variables categóricas, es decir, variables de texto

```python
# Data
countries = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [5000, 800, 900, 1000, 300]

# Gráfico
plt.bar(countries,population)
plt.show()
```

## Cambiando los colores y el tamaño

```python
plt.bar(countries,population,  width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

## Modificar los nombres de las variables en el eje ‘**X**’ con `xticks`

```python
plt.bar(countries,population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japón', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

## Crear un gráfico de manera horizontal con `barh`

```python
plt.barh(countries,population)
plt.show()
```

# Crear otro tipo de gráficas

## Histograma

```python
data = np.random.randint(1,50,100)

plt.hist(data)
plt.show()
```

Cambiar el estilo y el numero de barras

```python
plt.hist(data, bins=10, histtype='step') 
plt.grid(visible=None)
plt.show()
```

## Boxplot

```python
plt.boxplot(data)  #Muestra rango intercuartil y la mediana
plt.grid(visible=None)
plt.show()
```

Cambiar dirección, rellenar el rango inter cuartil `patch_artist` y enfocar la mediana `notch`

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.grid(visible=None)
plt.show()
```

## Scatter plot

```python
# Data
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)

# Plot
plt.scatter(x, y, s=area)
plt.grid(visible=None)
plt.show()
```

Cambiando el color y los marcadores

```python
plt.scatter(x,y, s=area, c= colors, marker = 'v') # cambiar el marker
plt.grid(visible=None)
plt.show()
```

Podemos agregar transparencias

```python
plt.scatter(x,y, s=area, c= colors, marker = 'o', alpha= 0.5) # transparencia
plt.grid(visible=None)
plt.show()
```
